```python
# extracts ['title','size','color','rating','review_text'] from page, saves as dictionary and stores it in a list
def get_reviews_from_page_v2(driver, data):
    #get all reviews in the current page
    results=driver.find_elements(by=By.CSS_SELECTOR,value='div[class="a-section celwidget"]')

    print('\t\tResult Number', len(results))
    i = 1
    for result in results: # for each review
        print(f'\t\t scraping review {i} on current page')
        title, size, color, rating, review_text = 'NA','NA','NA', 'NA','NA'
       
        try: # try to get the title
            title_box=result.find_element(by=By.CSS_SELECTOR,
                                          value='a[class="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"]')
            title=title_box.text.strip()
#             print(title)
        except NoSuchElementException as e: 
            print('\t\t could not extract title')
        
        try: # try to get size and color, both are in the same HTML
            size_color_box = result.find_element(by=By.CSS_SELECTOR,
                                         value='a[class="a-size-mini a-link-normal a-color-secondary"]')
            size_color_text = size_color_box.text # contains size and color as 'Size: sizeColor: color'
            text_parts = size_color_text.split('Color:') # splits into ['Size: size', 'color']
            if len(text_parts) == 2:
                size = text_parts[0].replace('Size:','').strip() # turns 'Size: size' to 'size'
                color = text_parts[1].strip() # gets color
        except NoSuchElementException as e:
            print('\t\t could not extract size or color')
    
        try: #try to get the ratings
            rating_box=result.find_element(by=By.CSS_SELECTOR,value='span[class="a-icon-alt"]')
            rating_text = rating_box.get_attribute('textContent') # gets 'X out of Y stars'
            rating = rating_text.split()[0].strip() # gets X from the string
#             print(f'\t rating: {rating}')
        except NoSuchElementException as e: 
            print('\t\t could not extract rating')
            
        try: # try to get the review_text
            review_text_box=result.find_element(by=By.CSS_SELECTOR,value='span[class="a-size-base review-text review-text-content"]')
            review_text=review_text_box.text.strip()
#             print(review_text)
        except NoSuchElementException as e:
            print('\t\t could not extract review_text')
            
        # add new review to dataframe
        current_review = {
            'title': title,
            'size': size,
            'color': color,
            'rating': rating,
            'review_text': review_text,
        }
        data.append(current_review)
#         print(f'\t\t wrote row {i}')
        i += 1
    # outputting all scraped reviews
    return data
```

```python
# goes through all pages in url to extract review info and write it into csv
# driver: selenium driver
# delay: number of seconds to wait
# urls: list of website url
# output_name: name of the ouput file of form filename.csv
def scrape_urls(driver, delay, urls, output_name):
    # create varaible to hold scraped information
#     scraped_reviews_df = pd.DataFrame(columns=['title','size','color','rating','review_text'])
    scraped_reviews = []
    
    # try go through each url
    try:
        # go through each url
        for i, url in enumerate(urls):
            # wait
            time.sleep(1)
            # open current page
            driver.get(url)
            print(f'Scraping url {i}')
            # handle cookie overlay
#             input() # not needed, these review pages don't have cookie overlay
            
            page_cnt = 1 # keeps track of page count
            
            while True: # keep going until there are no more pages
                print(f'\t page {page_cnt}')
                page_cnt += 1
                
                # extract reviews from current page and update scraped_reviews_df
                scraped_reviews = get_reviews_from_page_v2(driver, scraped_reviews)
                
                try: # try to go to next page
                    #wait until next buton appears
                    next_button = WebDriverWait(driver, delay).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'div[data-hook="pagination-bar"] li.a-last a')))

                    #check if the next button is disabled
                    if 'a-disabled' in next_button.get_attribute('class'):
                        print('no more pages')
                        break

                    # click the next button, increment page_cnt, and wait 
                    # next_button.click()
                    driver.execute_script("arguments[0].click();", next_button)
                    time.sleep(delay)

                except TimeoutException as e:
                    print('no more pages')
                    break
            # stop writing current review
            
    except Exception as e:
        print(f'An error occurred: {e}')
    print('Done scraping!')
    
    # turn list of dictionaries into dataframe
    scraped_reviews_df = pd.DataFrame(scraped_reviews,)
    # dropping any duplicates
    scraped_reviews_df_final = scraped_reviews_df.drop_duplicates()
    
    print(f'Exporting {output_name}')
    scraped_reviews_df_final.to_csv(output_name, index=False, encoding='utf-8-sig')
    print(f'Exported {output_name}')
```

## Additional Comments

We tried out best to deal with this by getting all the possible combinations for each filtered review section:  
* all stars, 5 star reviews, 4 star reviews, 3 star reviews, 2 star reviews, 1 star reviews, ciritical reviews
    * if a subsection of reviews had more than 100 reviews we included both top and most recent reviews

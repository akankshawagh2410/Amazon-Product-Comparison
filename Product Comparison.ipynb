## Opinion Mining

For opinion mining we made use of 'positive-words.txt' & 'negative-words.txt' for sentiment analysis

```python
#function that loads a lexicon of positive words to a set and returns the set
def loadLexicon(fname):
    newLex=set()
    lex_conn=open(fname)
    
    #add every word in the file to the set
    for line in lex_conn:
        newLex.add(line.strip())# remember to strip to remove the line-change character and white space
    lex_conn.close()

    return newLex
```

```python
# executes sentiment analysis
# output: mapping of a noun to its sentiments
#         - seniment is an array of the form [positive_count, negative_count, overall_count]
# posLex: lexicon of positively associated words
# negLex: lexicon of negatively associated words
# reviews: array of all reviews of form [[rating, review_string], ...]
def get_sentiments(posLex, negLex, reviews):
    # maps each noun to an array of form [positive, negative, overall_sentiment]
    noun_sentiment = {}
    # go through each review
    for review in reviews:
        rating, text = review
        # split review into sentences
        sentences = sent_tokenize(text)
        # go through each sentence and collect sentiments
        for sentence in sentences:
            # prepare sentence for sentiment analysis
            words = word_tokenize(sentence) # split sentence into words
            tagged_words = nltk.pos_tag(words) # tag words by part of speach
            nouns_in_sentence = set() # set of all nouns in the sentence
            # initialize variables to count sentiments
            positive_sentiment_count = 0
            negative_sentiment_count = 0
            
            # go through each word and find noun sentiments
            for tagged_word in tagged_words:
                # noun found
                if tagged_word[1].startswith('NN'):
                    noun = tagged_word[0].lower()
                    if len(noun) < 3: continue # ignore nouns with less than 3 characters
                    nouns_in_sentence.add(noun) # add noun to the set
                    
                # adjective found
                if tagged_word[1].startswith('JJ'):
                    if tagged_word[0] in posLex: # positive
                        positive_sentiment_count += 1
                    if tagged_word[0] in negLex: # negative
                        negative_sentiment_count += 1
            
            # getting overall feeling for words in the sentence
            overall_sentiment = positive_sentiment_count - negative_sentiment_count
            
            # update noun_sentiment
            for noun in nouns_in_sentence:
                if noun not in noun_sentiment: # noun does not already have a sentiment
                    # add noun and sentiment
                    noun_sentiment[noun] = [positive_sentiment_count, negative_sentiment_count, overall_sentiment]
                elif noun in noun_sentiment: # noun already has a sentiment
                    # update noun's sentiments
                    old_pos, old_neg, old_overall = noun_sentiment[noun] # getting previous sentiments
                    # calculating new sentiment
                    new_pos = old_pos + positive_sentiment_count
                    new_neg = old_neg + negative_sentiment_count
                    new_overall = new_pos - new_neg
                    # updating sentiment
                    noun_sentiment[noun] = [new_pos, new_neg, new_overall]
    # outputting sentiment analysis
    return noun_sentiment
```

```python
# returns nouns that have opnions that are either negative or positive
# output: one list of positively associated nouns and one list of negatively associated nouns
#        - each list is of the form [(noun, [positive_count, negative_count, overall_count]), ...]
#        - each list is sorted from the most of it's associated value to the least
# positive_words: path to textfile containing positive_words
# negative_words: path to textfile containing negative_words
# reviews: array of all reviews of form [[rating, review_string], ...]
def get_opinions(positive_words, negative_words, reviews):
    # check if positive_words goes to a valid file
    if not os.path.isfile(positive_words):
        return 'Error: positive_words does not go to a file'
    if positive_words.split('.')[-1] != 'txt':
        return 'Error: file in positive_words is not a textfile'
    # check if negative_words goes to a valid file
    if not os.path.isfile(negative_words):
        return 'Error: negative_words does not go to a file'
    if negative_words.split('.')[-1] != 'txt':
        return 'Error: file in negative_words is not a textfile'
    
    # load the positive and negative lexicons into sets
    posLex=loadLexicon(positive_words)
    negLex=loadLexicon(negative_words)
    
    # maps each noun to an array of form [positive, negative, overall_sentiment]
    noun_sentiment = get_sentiments(posLex,negLex, reviews)
    
    # get all positive and negative nouns in order from most to least positive/negative
    positive_words = [] # stores all positive nouns in descending order (most to least positive)
    negative_words = [] # stores all negative nouns in ascending order (most to least negative)
    
    # go through all words
    for noun in noun_sentiment.keys():
        curr_sentiment = noun_sentiment[noun] # get sentiment array
        curr_positive, curr_negative, curr_overall = curr_sentiment
        curr_info = (noun, curr_overall)
        # adding to appropriate array
        if curr_overall < 0: # overall negative
            negative_words.append(curr_info)
        if curr_overall > 0: # overall positive
            positive_words.append(curr_info)
    
    # sorting positive_words in descending order
    positive_words_sorted = sorted(positive_words, key=lambda x: x[1], reverse=True)
    # sorting negative_words in ascending order
    negative_words_sorted = sorted(negative_words, key=lambda x: x[1])
    
    return positive_words_sorted, negative_words_sorted
```

```python
# returns nouns that have differing opinions across all reviews, ranked by how often they show up
# output: mapping of noun to the reviews that have differing opinions of it
# positive_words: path to textfile containing positive_words
# negative_words: path to textfile containing negative_words
# reviews: array of all reviews of form [[rating, review_string], ...]
def get_differing_opinions(positive_words, negative_words, reviews):
    # check if positive_words goes to a valid file
    if not os.path.isfile(positive_words):
        return 'Error: positive_words does not go to a file'
    if positive_words.split('.')[-1] != 'txt':
        return 'Error: file in positive_words is not a textfile'
    # check if negative_words goes to a valid file
    if not os.path.isfile(negative_words):
        return 'Error: negative_words does not go to a file'
    if negative_words.split('.')[-1] != 'txt':
        return 'Error: file in negative_words is not a textfile'
    
    # load the positive and negative lexicons into sets
    posLex=loadLexicon(positive_words)
    negLex=loadLexicon(negative_words)
    
    # mapping of noun to array storing the differing opinions
    differing_opinions = {}
    
    # go through upper triangle of review comparision diagonal matrix
    for i in range(len(reviews)):
        # getting sentiment analysis for current review
        review1_sentiments = get_sentiments(posLex, negLex, [reviews[i]])
        review1_nouns = review1_sentiments.keys()
        # compare review[i] to all reviews coming after it
        for j in range(i+1, len(reviews)):
            # getting sentiment analysis for secondary review
            review2_sentiments = get_sentiments(posLex, negLex, [reviews[j]])
            review2_nouns = review2_sentiments.keys()
            
            # find what nouns both reviews share
            shared_nouns = [noun for noun in review1_nouns if noun in review2_nouns]
            # see if they feel differently about the noun
            for noun in shared_nouns:
                # getting individual review opinions
                r1_pos, r1_neg, r1_overall = review1_sentiments[noun] # info for review 1
                r2_pos, r2_neg, r2_overall = review2_sentiments[noun] # info for review 2
                # normalizing overall sentiments
                r1_opinion = 1 if r1_overall > 0 else (-1 if r1_overall < 0 else 0)
                r2_opinion = 1 if r2_overall > 0 else (-1 if r2_overall < 0 else 0)
                # seeing if opinions are different
                if r1_opinion != r2_opinion:
                    r1_index = i
                    r2_index = j
                    # making review info for review 1
                    review1 = {
                        'review_index': r1_index,
                        'overall_opinion': r1_opinion,
                        'breakdown': (noun, [r1_pos, r1_neg, r1_overall]),
                    }
                    # making review info for review 2
                    review2 = {
                        'review_index': r2_index,
                        'overall_opinion': r2_opinion,
                        'breakdown': (noun, [r2_pos, r2_neg, r2_overall]),
                    }
                    review_pair = (review1, review2)
                    # save current review_pair
                    if noun not in differing_opinions: # add to differing_opinions
                        differing_opinions[noun] = [review_pair]
                    elif noun in differing_opinions: # update differing_opinions
                        current_reviews = differing_opinions[noun]
                        current_reviews.append(review_pair)
                        differing_opinions[noun] = current_reviews
        # output differing_opinions
        return differing_opinions
```

## Topic Modeling

```python
# stop words used in topic modeling
sw = set(stopwords.words('english'))
```

__tokenizer()__

```python
# convert each review to a list of words
# output: list of all words
# reviews_text: review_string
# stop_words: set of stop words
def tokenizer(review_text, stop_words):
    return [word for word in word_tokenize(review_text.lower()) if word not in stop_words and len(word)>2]
```

__clean_reviews()__

```python
# cleans reviews and outputs a 2d list, where the inner list is all the words in each review
# output: array of form [[<all words in the review with the rating added to the end>], ...]
# reviews: array of all reviews of form [[rating, review_string], ...]
# stop_words: set of stop words for the tokenizer
def clean_reviews(reviews, stop_words):
    cleaned_reviews = [] # of cleaned reviews
    # go through each review
    for review_info in reviews:
        rating, review_text = review_info
        # clean and tokenize review_text
        cleaned_review_text = tokenizer(review_text.lower(), stop_words)
        # add rating
        rating_str = str(rating).strip()
        cleaned_review_text.append(rating_str)
        # add to list of cleaned reviews
        cleaned_reviews.append(cleaned_review_text)
    return cleaned_reviews
```

__train_lda()__

```python
# trains lda model for k groups
# output: trained model
# reviews: array of all reviews of form [[<all words in the review with the rating added to the end>], ...]
# k: the number of topics to be made
# n: number to determine number of words to get from each topic
# max_iter: maximum number of iterations to train the model for
# step: number of training iterations to perform each time
def train_lda(reviews, k, max_iter, step):
    # initialize model
    reviews_model = tp.LDAModel(k=k)
    # load reviews
    for review in reviews:
        reviews_model.add_doc(review)
    # train model
    print(f'\t Training LDA for k={k}')
    for i in range(0, max_iter, step):
        reviews_model.train(step)
#     print(f'\t Done training LDA for k={k}')
    # output trained model
    return reviews_model
```

__make_topics()__

```python
# trains LDA model and returns list of topics for best trained model
# output: list of tuples of form [(topic_number,topic_list), ...]
# reviews: array of all reviews of form [[rating, review_string], ...]
# min_k: the minimum number of topics to be made
# max_k: the maximum number of topics to be made, inclusive
# stop_words: set of stop words to be used in tokenization
# n: number to determine number of words to get from each topic
# max_iter: maximum number of iterations to train the model for
# step: number of training iterations to perform each time
def make_topics(reviews, min_k, max_k, stop_words, n, max_iter=500, step=10):
    # intialize optimization variables
    best_model = None # stores the best model
    best_k = None # stores the k value of the best model
    best_perplexity = float('-inf') # this is the metric to evaluate the best model
    
    # process data to be put into model
    print('Processing data')
    cleaned_reviews = clean_reviews(reviews, stop_words)
    print('Data processed')
    
    print('Beginning model training and optimization')
    # train and optimize model
    for i in range(min_k, max_k+1):
        curr_model = train_lda(cleaned_reviews, i, max_iter, step)
        # getting perplexity
        curr_perplexity = curr_model.perplexity
        # checking whether or not to save model
        if curr_perplexity > best_perplexity:
            # current model is best model, save
            best_model = curr_model
            best_k = i
            best_perplexity = curr_perplexity
    print(f'optimized LDA model trained')
    
    # array to save k topics
    k_topics = []
    print('Creating topics')
    # get top n words for each topic
    for i in range(best_model.k):
        curr_top_words = [pair[0] for pair in best_model.get_topic_words(i, top_n=n)]
        curr_topic_pair = (i, curr_top_words)
        k_topics.append(curr_topic_pair)
    print('Topics created')
    print('Done!')
    # output k_topics
    return k_topics
```

__make_k_topics()__

```python
# make k topics for product reviews and outputs the k topics with top n number of words
# output: list of tuples of form [(topic_number,topic_list), ...]
# reviews: array of all reviews of form [[rating, review_string], ...]
# k: number of topics to be made
# stop_words: set of stop words to be used in tokenization
# n: number to determine number of words to get from each topic
# max_iter: maximum number of iterations to train the model for
# step: number of training iterations to perform each time
def make_k_topics(reviews, k, stop_words, n, max_iter=500, step=10):
    # process data to be put into model
    print('Processing data')
    cleaned_reviews = clean_reviews(reviews, stop_words)
    print('Data processed')
    
    # initialize LDA model
    reviews_model = tp.LDAModel(k=k)
    print(f'LDA model initialized')
    
    # load data into model
    for review in cleaned_reviews:
        reviews_model.add_doc(review)
        
    # train LDA model
    print(f'training LDA model')
    for i in range(0, max_iter, step):
        reviews_model.train(step)
    print('LDA model trained')
    
    # array to save k topics
    k_topics = []
    print('Creating topics')
    # get top n words for each topic
    for i in range(reviews_model.k):
        curr_top_words = [pair[0] for pair in reviews_model.get_topic_words(i, top_n=n)]
        curr_topic_pair = (i, curr_top_words)
        k_topics.append(curr_topic_pair)
    print('Topics created')
    print('Done!')
    # output k_topics
    return k_topics
```

__print_topics()__

```python
# prints out the result of make_k_topics
# k_topics: result of make_k_topics
#           - form [(topic_number, topic_list), ...]
def print_topics(k_topics):
    k = len(k_topics) # get k
    for i in range(k):
        curr_topic_number = k_topics[i][0]
        curr_topic_words = k_topics[i][1]
        print(f'{curr_topic_number}: {curr_topic_words}')
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Create sentiment analyzer 
vader = SentimentIntensityAnalyzer()
```

```python
df1['scores'] = df1['review_text'].apply(lambda x: vader.polarity_scores(x) if isinstance(x, str) else {'compound': 0.0})
df1['compound'] = df1['scores'].apply(lambda d: d['compound'])
df1['sentiment'] = df1['compound'].apply(lambda c: 'pos' if c >= 0 else 'negative') 

# Apply sentiment analysis for product 1
#df1['scores'] = df1['review_text'].apply(lambda x: vader.polarity_scores(x))
#df1['compound'] = df1['scores'].apply(lambda d: d['compound'])
#df1['sentiment'] = df1['compound'].apply(lambda c: 'pos' if c>=0 else 'negative')
```

```python
# Apply sentiment analysis for product 2 
df2['scores'] = df2['review_text'].apply(lambda x: vader.polarity_scores(x) if isinstance(x, str) else {'compound': 0.0})
df2['compound'] = df2['scores'].apply(lambda d: d['compound']) 
df2['sentiment'] = df2['compound'].apply(lambda c: 'pos' if c>=0 else 'negative')
```

```python
# Divide reviews into positive and negative
pos_reviews_p1 = df1[df1['sentiment'] == 'pos']
neg_reviews_p1 = df1[df1['sentiment'] == 'negative']

pos_reviews_p2 = df2[df2['sentiment'] == 'pos']  
neg_reviews_p2 = df2[df2['sentiment'] == 'negative']
```

```python
print("Stats for Product 1:")
print(f"- Number of positive reviews: {len(pos_reviews_p1)}") 
print(f"- Number of negative reviews: {len(neg_reviews_p1)}")

print("Stats for Product 2:")
print(f"- Number of positive reviews: {len(pos_reviews_p2)}")
print(f"- Number of negative reviews: {len(neg_reviews_p2)}")
```

Here is a comparative report analyzing how the 2 products are different based on the sentiment analysis and word clouds:

__Comparative Sentiment Analysis Report__

__I. Product Review Statistics:__

Product 1 has more positive sentiment overall out of which 63% of the reviews were positive and 37% were negative.

Product 2 has more mixed sentiment in which 55% were positive reviews and 45% were negative reviews.

So while both products had a majority of positive reviews, Product 1 reviews were even more overwhelmingly positive compared to Product 2.


__II. Frequent Review Words:__

Frequent Positive Review Words:

Product 1's top positive words focused around comfort, style/looks, warmth, quality and fit.
Product 2 also mentioned warmth, comfort and quality as positives but not as frequently.
Unique positives highlighted for Product 2 were traction and lightweight.
So both products deliver on comfort, warmth and quality but Product 1 seems better rated for style and fit by customers.

Frequent Negative Review Words:

Product 1 negatives revolved around issues with sizing/fit or depth not being enough for warmth.
Product 2 top negatives included traction problems, waterproofing issues, sizing inaccuracies and cheap feel.
Product 2 seems to fall short for some customers on core functionality - traction, waterproofing ability and build quality leading to more negative sentiment.

__III. Key Differences:__

Product 1
- Much more positive sentiment
- Style and looks were a strong plus
- Found comfortable and warm by most users

Product 2
- More mixed reviews
- Fell short on traction and waterproofing capability
- Sizing inconsistencies and feel/quality disappointed some


__IV. SUMMARY:__

In brief, although both products fulfilled customer expectations regarding warmth and comfort, Product 1 received higher ratings for its style and consistent sizing. On the other hand, Product 2 faced core functionality issues, particularly receiving more negative reviews concerning traction and waterproofing capabilities.
